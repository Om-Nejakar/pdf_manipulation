```python
# extracting the text from the pdf 
from PyPDF2 import PdfReader

reader = PdfReader("pdf1.pdf") # creating and reader object
number_of_pages = len(reader.pages)
page = reader.pages[0]
text = page.extract_text()
print(reader.pages) #its an object with virtual_list (lazy loading only loads page when required) internally calculate the no. of pages
print(number_of_pages)
print(text)
```

```python
# extracting the meta-data from the pdf
from PyPDF2 import PdfReader as pdr 
with open("pdf2.pdf", "rb") as pdf_file:
    reader = pdr(pdf_file)

    metadata = reader.metadata
    print(metadata)
    print(metadata.author)
    print(metadata.title)
    print(metadata.creation_date)
```

```python
# merging the pdfs 
from PyPDF2 import PdfReader, PdfWriter

pdf_writer = PdfWriter() # its save all the accumulated pages
for pdf_file in ['pdf1.pdf', 'pdf2.pdf']:
    reader = PdfReader(pdf_file)
    for page in reader.pages:
        pdf_writer.add_page(page) 
with open('merged.pdf', 'wb') as output_file: #opening the file in rb/wb (because pdf store value in binary and no data gets corrupted and its safe)
    pdf_writer.write(output_file) # that accumulated pages are written in outputfile
```

```python
#splitting the pdfs
from PyPDF2 import PdfReader, PdfWriter
writer1 = PdfWriter()
writer2 = PdfWriter()

with open("zscaler.pdf","rb") as input_file:
    reader = PdfReader(input_file)
    for i in range(100):
        writer1.add_page(reader.pages[i])
    for j in range(101,225):
        writer2.add_page(reader.pages[j])

writer1.encrypt('pass@123') #password encryption 

with open("part1.pdf", "wb") as output_file1:
    writer1.write(output_file1)
with open("part2.pdf", "wb") as output_file2:
    writer2.write(output_file2)
```

```python
#adding watermarks to the pdf
from PyPDF2 import PdfReader, PdfWriter

with open('pdf1.pdf','rb') as main_pdf, open('watermark.pdf', 'rb') as watermark_pdf:
    reader = PdfReader(main_pdf)
    watermark_reader = PdfReader(watermark_pdf)

    writer = PdfWriter()
    watermark_page = watermark_reader.pages[0]

    for page in reader.pages:
        page.merge_page(watermark_page)
        writer.add_page(page)
    
    with open('watermarked.pdf', 'wb') as output_file:
        writer.write(output_file)
```
